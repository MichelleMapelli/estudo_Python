### Agora irei criar um DataFrame para armazenar os 3 cálculos feitos acima

```python
df_resultado_descinst = pd.DataFrame({
    
    'contagem_nome':df_contagem_nome,
    'mediana_remuneracao':df_mediana_descinst,
    'soma_remuneracao':df_soma_descinst,
    'tkm_servidor':df_tkm_servidores
})

df_resultado_descinst = df_resultado_descinst.sort_values(by='soma_remuneracao', ascending=False)
df_resultado_descinst.head(5)
```

```python
# Selecionar os 10 primeiros resultados
top_10 = df_resultado_descinst.head(10)

#Definir uma paleta de cores de azul
paleta = sns.color_palette("Blues", len(top_10))

# Configurar o gráfico de barras
plt.figure(figsize=(12, 8))
plt.bar(top_10.index, top_10['soma_remuneracao'], color=paleta)

# Adicionar títulos e rótulos
plt.title('Top 10 Instituições por Soma de Remuneração', fontsize=14, fontweight='bold')
plt.xlabel('Instituição', fontsize=12)
plt.ylabel('Soma da Remuneração', fontsize=12)

# Aumentar o limite superior do eixo y
plt.ylim(0, top_10['soma_remuneracao'].max() * 1.1)

# Adicionar rótulos nas barras
for i, value in enumerate(top_10['soma_remuneracao']):
    plt.text(i, value + 10000 , f'{value:.2f}', ha='center', fontsize=10)
    
# Aprimirar Layout
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Mostrar gráfico
```

```python
# Ordenar pelo valor contagem_nome de forma decrescente
df_resultado_descinst_contagem = df_resultado_descinst.sort_values(by='contagem_nome', ascending=False)

# Selecionar os top 10 maiores quantidades de servidores recebendo em Abril/2024
top_10_ser = df_resultado_descinst_contagem.head(10)

#Configurar gráfico
plt.figure(figsize=(12,8))
plt.bar(top_10_ser.index, top_10_ser['contagem_nome'], color=paleta)

# Adicionar títulos e rótulos
plt.title('Top 10 Instituições por quantidade de Servidores recebendo em Abril de 2024', fontsize=14, fontweight='bold')
plt.xlabel('Instituição', fontsize=12)
plt.ylabel('Contagem de Servidores', fontsize=12)

# Aumentar o limite superior do eixo y
plt.ylim(0, top_10_ser['contagem_nome'].max() * 1.1)

# Adicionar rótulos nas barras
for i, value in enumerate(top_10_ser['contagem_nome']):
    plt.text(i, value + 1000, f'{value:.0f}', ha='center', fontsize=10)
    
# Aprimirar o Layout
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Apresentar gráfico
plt.show()
```
