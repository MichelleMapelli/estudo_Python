```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import pandas as pd

# Caminho para o arquivo Excel
file_path = 'servidores-2024-04.xlsx'

try:
    # Ler o arquivo Excel
    df = pd.read_excel(file_path)

    # Visualizar os primeiros registros
    print(df.head())
except UnicodeDecodeError as e:
    print(f"Erro de decodificação: {e}")
except FileNotFoundError:
    print("O arquivo não foi encontrado.")
except ValueError as e:
    print(f"Erro ao processar o arquivo Excel: {e}")
except Exception as e:
    print(f"Ocorreu um erro inesperado: {e}")
```

```python
df.head(5)
```

```python
df.describe()
```

```python
colunas = df.columns
print(colunas)
```

```python
df_mediana = df.groupby('descinst')['remuner'].median()
```

```python
df_soma = df.groupby('descinst')['remuner'].sum()
```

```python
df_contagem_nome = df.groupby('descinst')['nome'].count()
print(df_contagem_nome)
```

```python
df_resultado = pd.DataFrame({

                             'contagem_nome':df_contagem_nome,
                             'soma':df_soma,
                             'mediana_remuner':df_mediana
})

df_resultado = df_resultado.sort_values(by='soma', ascending=False)
df_resultado.head(15)
```
