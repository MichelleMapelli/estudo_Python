### Eu quero saber o nome das colunas, pois como são muitas, não consigo ver acima

```python
colunas = df.columns
print(colunas)
```

```python
df_mediana_descinst = df.groupby('descinst')['remuner'].median()
print(df_mediana_descinst)
```

```python
df_soma_descinst = df.groupby('descinst')['remuner'].sum()
print(df_soma_descinst)
```

```python
df_contagem_nome = df.groupby('descinst')['nome'].count()
print(df_contagem_nome)
```

### Agora irei criar um DataFrame para armazenar os 3 cálculos feitos acima

```python
df_resultado_descinst = pd.DataFrame({
    
    'contagem_nome':df_contagem_nome,
    'mediana_remuneracao':df_mediana_descinst,
    'soma_remuneracao':df_soma_descinst
})

df_resultado_descinst = df_resultado_descinst.sort_values(by='soma_remuneracao', ascending=False)
df_resultado_descinst.head(5)
```
